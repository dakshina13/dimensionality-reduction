```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
```

```python
data = pd.read_csv('data3.csv',index_col=False)
data=data.drop('Unnamed: 32',axis=1)
data.head()
```

```python
data.isnull().sum()
```

```python
X = data.copy()
y = data['diagnosis']
```

```python
X=X.drop('diagnosis',axis=1)
X=X.drop('id',axis=1)
X
```

```python
target=data.diagnosis
y=[]
for t in target:
    if t=="M":
        y.append(1)
    else:
        y.append(0)
y=np.array(y)
y
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 43)
```

```python
y_test
```

```python
sel = SelectFromModel(LogisticRegression(penalty = 'l2', C = 0.05, solver = 'liblinear'))
sel.fit(X_train, y_train)
selectedArray=sel.get_support(indices=True)
selectedArray=np.array(selectedArray)
print(selectedArray)
selectedArray.shape
```

```python
features=[feat for feat in data]
features.remove('id')
features.remove('diagnosis')
for s in selectedArray:
    print(features[s])
```

```python
coeffArray=sel.estimator_.coef_
coeffArray=np.array(coeffArray)
featureCoeffMap={}
for i,f in enumerate(features):
    featureCoeffMap[f]=coeffArray[0][i]
for s in selectedArray:
    print(coeffArray[0][s])
```

```python
import json
print(json.dumps(featureCoeffMap,sort_keys=True, indent=4))
```

```python
X_train_l1 = sel.transform(X_train)
X_test_l1 = sel.transform(X_test)
```

```python
def run_randomForest(X_train, X_test, y_train, y_test):
    clf = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
    clf = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print('Accuracy: ', accuracy_score(y_test, y_pred))
```

```python
%%time
run_randomForest(X_train_l1, X_test_l1, y_train, y_test)
```
