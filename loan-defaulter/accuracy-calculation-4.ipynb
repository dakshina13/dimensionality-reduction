# Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("dataset/application_data.csv")
```

```python
train
```

```python
features=[feat for feat in train]
features
```

```python
train=train.dropna()
train
```

## Label Encoding

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
```

```python
data_type_list=train.dtypes
type(data_type_list)
```

```python
print("Printing the list of non float columns")
for k,v in train.dtypes.items():
    if v not in ["int32","int64","float64"]:
        print(k,v)
```

```python
print("Converting the labels")
for k,v in train.dtypes.items():
    if v not in ["int32","int64","float64"]:
        print("For k "+k)
        print(train[k].unique())
        train[k]=label_encoder.fit_transform(train[k])
        print(train[k].unique())
        
```

```python
print("Checking if any column has string")
for k,v in train.dtypes.items():
    if v not in ["int32","int64","float64"]:
        print(k,v)
```

## Removing NaN values from data frame

```python
print("Droping target and id column")
df=train.drop("TARGET",axis=1)
df=df.drop("SK_ID_CURR",axis=1)
df
```

```python
target=train.TARGET
```

# Model with all features

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train_m1, X_test_m1, y_train_m1, y_test_m1 = train_test_split(df, target, test_size = 0.3)
lgr=LogisticRegression(penalty = 'l2', C = 0.05, solver = 'liblinear')
```

```python
%%time
lgr.fit(X_train_m1,y_train_m1)
```

```python
lgr.score(X_test_m1,y_test_m1)
```
