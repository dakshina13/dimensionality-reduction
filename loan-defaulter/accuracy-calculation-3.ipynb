## Label Encoding

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
```

```python
data_type_list=train.dtypes
type(data_type_list)
```

```python
print("Printing the list of non float columns")
for k,v in train.dtypes.items():
    if v not in ["int32","int64","float64"]:
        print(k,v)
```

```python
print("Converting the labels")
for k,v in train.dtypes.items():
    if v not in ["int32","int64","float64"]:
        print("For k "+k)
        print(train[k].unique())
        train[k]=label_encoder.fit_transform(train[k])
        print(train[k].unique())
        
```

```python
print("Checking if any column has string")
for k,v in train.dtypes.items():
    if v not in ["int32","int64","float64"]:
        print(k,v)
```

## Removing NaN values from data frame

```python
null_list=[]
for k in train:
    if train[k].isnull().any():
        null_list.append(k)
        train[k].fillna(train[k].median(),inplace=True)
null_list=np.array(null_list)
null_list.shape
```

```python
print("Droping target and id column")
df=train.drop("TARGET",axis=1)
df=df.drop("SK_ID_CURR",axis=1)
df
```

```python
target=train.TARGET
```

# Forward selection using SFS

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.linear_model import LogisticRegression as LGR
```

```python
forword_selection = SFS(
           LGR(max_iter=1000),
           k_features='best', 
           forward=True, 
           floating=False, 
           verbose=2,
           scoring='accuracy',  
           cv=0)

forword_selection = forword_selection.fit(df, target)
```

```python
# Lets access the indices of the best features directly via the k_feature_idx_ attribute:
forword_selection.k_feature_names_, forword_selection.k_feature_idx_
```
